```python
df1 = pd.DataFrame({
    'Item': ['TV', 'TV', 'TV', 'AC'],
    'Company': ['LG', 'VIDEOCON', 'LG', 'SONY'],
    'Rupees': [12000, 10000, 15000, 14000],
    'USD': [700, 650, 800, 750]
})
df1
```

```python
df1.loc[4] = ['Bag','Skybags',6500,75]
print(df1)
```

```python
max_price = df1[(df1['Company'] == 'LG') & (df1['Item'] == 'TV')]['Rupees'].max()
print("\nMaximum price of LG TV:", max_price)
```

```python
sum_products = df1['Rupees'].sum()
print("\nSum of all products:", sum_products)
```

```python
sony = df1[df1['Company'] == 'SONY']['USD'].median()
print("\nMedian USD price of Sony products:", sony)
```
